```python
from MCTS import MCTS
from Node import Node
from Game import TicTacToe_Board
from random import random
```

```python
# Random policy evaluation to check MCTS
def NN(state):
    p = []
    for elem in state:
        p.append(random())
    v = random()
    return p,v
```

```python
#MCTS game:
T = 1
state = [0]*9
end = False
player = -1

game = TicTacToe_Board()
root = Node(state, 1, player)
mcts = MCTS(game, root, NN, ngames=100)

node = root
while not end:
    mcts.explore(node)
    a = mcts.play(node, T)
    node = node.children[a]
    game.plot(node.state)
    
    end, winner = game.check_end(node.state)
```
